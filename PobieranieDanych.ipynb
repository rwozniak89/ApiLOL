```python
def get_matches_details(maches_list, max_count=0):
    from dateutil.relativedelta import relativedelta

    import time
    def convertSecToTime(seconds):
        return time.strftime("%H:%M:%S", time.gmtime(seconds))
    
    def timeIsBetween(t, t_from,t_thru):
        if(t is None):
            return False
        elif(t.hour >= t_from and t.hour <= t_thru):
            return True
        else:
            return False
        
    def dayIsSchoolDay(d):
        if(d is None):
            return False
        elif(d.isoweekday() >= 1 and d.isoweekday() <= 5):
            return True
        else:
            return False
        
    def countGameOnDay(dt_temp, dt_game, counter):
        if(dt_game is None):
            return 0, None
        if(dt_temp is None):
            dt_temp = dt_game.date()
            counter = 1
        elif (dt_temp == dt_game.date()):
            counter = counter + 1
        else :
            dt_temp = dt_game.date()
            counter = 1
        return dt_temp, counter 

    mat_list = []
    
    gameOnDay_date_start = None
    gameOnDay_count_start = 0
    gameOnDay_date_end = None
    gameOnDay_count_end = 0
    
    gameOnDay_end = 0
    for index, val in enumerate(maches_list):
        match_id = val
        get_match_str=f'https://europe.api.riotgames.com/lol/match/v5/matches/{match_id}?api_key={key}'


        test_iterator = 0
        while test_iterator <=15:
            get_match_response =requests.get(get_match_str)
            if(get_match_response.status_code != 200):
                print(f'code {get_match_response.status_code}, status fail... wait 10 sek, try {test_iterator}')
                time.sleep(10)
                test_iterator = test_iterator + 1

            elif (get_match_response.status_code == 200):
                test_iterator = 0
                print(f'{index} index, code: {get_match_response.status_code}, status Ok!')

                match = get_match_response.json()

                indexOf2 = -1
                isFather = False
                try:
                    indexOf2 = match['metadata']['participants'].index(puuid2)
                    if(indexOf2 > -1): isFather = True
                except:
                    #indexOf2 = -1 
                    isFather = False
                    #print("Player2 is off") 
                    
                indexOf3 = -1
                isPlayer3 = False
                try:
                    indexOf3 = match['metadata']['participants'].index(puuid3)
                    if(indexOf3 > -1): isPlayer3 = True
                except:
                    #indexOf2 = -1 
                    isPlayer3 = False
                    #print("Player2 is off") 

                winner = False

                indexOfTeam = match['metadata']['participants'].index(puuid)
                lenOfTram = len(match['metadata']['participants'])

                isFirstTeam = (indexOfTeam < lenOfTram/2)
                if(isFirstTeam == True):
                    winner = match['info']['teams'][0]['win']
                else: 
                    winner = match['info']['teams'][1]['win']

                addData = relativedelta(hours=2)

                import pandas as pd
                dt1 = pd.to_datetime(match['info']['gameCreation'], unit='ms').to_pydatetime()
                dt1 = dt1  + addData
                
                gameOnDay_date_start, gameOnDay_count_start = countGameOnDay(gameOnDay_date_start, dt1, gameOnDay_count_start)
                
                if "gameEndTimestamp" in match['info']:
                    dt2 = pd.to_datetime(match['info']['gameEndTimestamp'], unit='ms').to_pydatetime()
                    dt2 = dt2  + addData
                    gameEndTimestampDT = dt2.strftime("%Y-%m-%d, %H:%M:%S")
                    gameEndTimestampD = dt2.strftime("%Y-%m-%d")
                    gameEndTimestampT = dt2.strftime("%H:%M:%S")
                    gameOnDay_date_end, gameOnDay_count_end = countGameOnDay(gameOnDay_date_end, dt2, gameOnDay_count_end)
                else:
                    dt2 = None
                    gameEndTimestampDT = 0
                    gameEndTimestampD = 0
                    gameEndTimestampT = 0
                
  
                dt3 = pd.to_datetime(match['info']['gameStartTimestamp'], unit='ms').to_pydatetime()
                dt3 = dt3 + addData

                mat = dict()
                mat = {
                    'id' : match_id,
                    'winner' : winner,
                    'player1_name': name,
                    'isPlayer2' : isFather,
                    'player2_name': name2,
                    'isPlayer3' : isPlayer3,
                    'player3_name': name3,
                    'gameOnDay_start' : gameOnDay_count_start,
                    'gameStartDT' : (dt3.strftime("%Y-%m-%d, %H:%M:%S")),
                    'gameStartD' : (dt3.strftime("%Y-%m-%d")),
                    'gameStartT' : (dt3.strftime("%H:%M:%S")),
                    'gameStartD-SchoolDay' : (dayIsSchoolDay(dt3)),
                    'gameStartT-21-23' : (timeIsBetween(dt3,21,23)),
                    'gameStartT-22-23' : (timeIsBetween(dt3,22,23)),
                    'gameStartT-0-1' : (timeIsBetween(dt3,0,1)),
                    'gameStartT-2-3' : (timeIsBetween(dt3,2,3)),
                    'gameStartT-4-5' : (timeIsBetween(dt3,4,5)),
                    'gameStartT-6-8' : (timeIsBetween(dt3,6,8)),
                    'gameStartT-0-3' : (timeIsBetween(dt3,0,3)),
                    'gameStartT-0-4' : (timeIsBetween(dt3,0,4)),
                    'gameStartT-0-5' : (timeIsBetween(dt3,0,5)),
                    'gameStartT-0-8' : (timeIsBetween(dt3,0,8)),
                    'gameOnDay_end' : gameOnDay_count_end,
                    'gameEndDT' : gameEndTimestampDT,
                    'gameEndD' : gameEndTimestampD,
                    'gameEndT' : gameEndTimestampT,
                    'gameEndD-SchoolDay' : (dayIsSchoolDay(dt2)),
                    'gameEndT-21-23' : (timeIsBetween(dt2,21,23)),
                    'gameEndT-22-23' : (timeIsBetween(dt2,22,23)),
                    'gameEndT-0-1' : (timeIsBetween(dt2,0,1)),
                    'gameEndT-2-3' : (timeIsBetween(dt2,2,3)),
                    'gameEndT-4-5' : (timeIsBetween(dt2,4,5)),
                    'gameEndT-6-8' : (timeIsBetween(dt2,6,8)),
                    'gameEndT-0-3' : (timeIsBetween(dt2,0,3)),
                    'gameEndT-0-4' : (timeIsBetween(dt2,0,4)),
                    'gameEndT-0-5' : (timeIsBetween(dt2,0,5)),
                    'gameEndT-0-8' : (timeIsBetween(dt2,0,8)),
                    'gameDurationS': (match['info']['gameDuration']),
                    'gameDurationT': convertSecToTime(match['info']['gameDuration']),
                    'gameCreationDT' : (dt1.strftime("%Y-%m-%d, %H:%M:%S")),
                    'gameCreationD' : (dt1.strftime("%Y-%m-%d")),
                    'gameCreationT' : (dt1.strftime("%H:%M:%S")),
                    'gameMode': match['info']['gameMode'],
                    'gameName': match['info']['gameName'],
                    'gameType': match['info']['gameType'],
                    'player1_id': puuid,
                    'player2_id': puuid2,
                }
                mat_list.append(mat)
                break
        if (max_count != 0 and index >= max_count):
            break
    return mat_list
```
